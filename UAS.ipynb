```python
from scipy.stats import shapiro

# Memilih kolom yang akan diuji normalitas, misalnya 'Produksi'
kolom_uji = ['Produksi', 'Luas Panen',	'Curah hujan',	'Kelembapan',	'Suhu rata-rata']


# Menentukan apakah data berdistribusi normal atau tidak
alpha = 0.05
for i in kolom_uji :
    # Melakukan uji normalitas menggunakan Shapiro-Wilk
    stat, p_value = shapiro(data[i])

    # Menampilkan hasil uji
    print(f'Statistik Uji: {stat}\nP-Value: {p_value}')

    if p_value > alpha:
        print(f'Data {i} berdistribusi normal (tidak dapat menolak H0)')
    else:
        print(f'Data {i} tidak berdistribusi normal (menolak H0)')
```

```python
from scipy.stats import kruskal

# Membuat kelompok data berdasarkan Provinsi
grouped_data = [data['Produksi'][data['Provinsi'] == provinsi] for provinsi in data['Provinsi'].unique()]

# Melakukan uji Kruskal-Wallis
statistic, p_value = kruskal(*grouped_data)

# Menampilkan hasil uji
print(f'Statistik Uji Kruskal-Wallis: {statistic}\nP-Value: {p_value}')

# Menentukan apakah hasil uji signifikan
alpha = 0.05
if p_value < alpha:
    print('Terdapat perbedaan signifikan antara setidaknya dua kelompok.')
else:
    print('Tidak terdapat perbedaan signifikan antara kelompok.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import seaborn as sns


# Memilih fitur yang akan digunakan untuk clustering
features = data[['Produksi', 'Luas Panen', 'Curah hujan', 'Kelembapan', 'Suhu rata-rata']]

# Standarisasi fitur
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Menentukan jumlah kluster dengan metode elbow
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Visualisasi metode elbow
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Metode Elbow untuk Menentukan Jumlah Kluster')
plt.xlabel('Jumlah Kluster')
plt.ylabel('Inertia')
plt.show()

# Berdasarkan elbow plot, kita pilih jumlah kluster yang optimal
optimal_k = 3

# Melakukan clustering dengan K-Means
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)

# Visualisasi hasil clustering
sns.pairplot(data, hue='Cluster')
plt.suptitle('Pairplot dengan Warna Berdasarkan Cluster', y=1.02)
plt.show()

# Menampilkan dataframe dengan kolom hasil clustering
data
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Menggunakan PCA untuk mereduksi dimensi menjadi 2
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)
data['PCA1'] = pca_result[:, 0]
data['PCA2'] = pca_result[:, 1]

# Visualisasi hasil kluster dengan PCA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=data, palette='viridis', s=100, alpha=0.8)
plt.title('Visualisasi Hasil Kluster dengan PCA')
plt.xlabel('Principal Component 1 (PCA1)')
plt.ylabel('Principal Component 2 (PCA2)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


# Extract production data for the province you are interested in (replace "Aceh" with your province)
province_data = data[data['Provinsi'] == 'Aceh']['Produksi'].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler()
province_data_normalized = scaler.fit_transform(province_data)

# Create sequences for training the LSTM model
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        seq = data[i:i+seq_length]
        label = data[i+seq_length:i+seq_length+1]
        sequences.append((seq, label))
    return np.array(sequences)

seq_length = 3  # You can adjust this based on your preference
sequences = create_sequences(province_data_normalized, seq_length)

# Split the data into features and labels
X = sequences[:, 0]
y = sequences[:, 1]

# Convert data to TensorFlow-compatible format
X = np.array([x.reshape((seq_length, 1)).astype(np.float32) for x in X])
y = np.array([yi.astype(np.float32) for yi in y])

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(seq_length, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X, y, epochs=100, verbose=0)

# Predict the next year's production
future_steps = 1
future_predictions = []
last_sequence = province_data_normalized[-seq_length:].reshape((1, seq_length, 1))
for i in range(future_steps):
    pred = model.predict(last_sequence)
    future_predictions.append(pred)
    last_sequence = np.append(last_sequence[:, 1:, :], pred.reshape((1, 1, 1)), axis=1)

# Inverse transform the predictions to get the actual values
future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
```

```python
import numpy as np
import pandas as pd

# Assuming you have the LSTM model and scaler already defined

# Number of steps to predict into the future
future_steps = 5  # You can adjust this based on your requirements

# Function to predict next year's production for a given province
def predict_next_year(model, scaler, province_data):
    seq_length = model.input_shape[1]
    
    # Normalize the data
    province_data_normalized = scaler.transform(province_data.reshape(-1, 1))

    # Use the last sequence in the data for prediction
    last_sequence = province_data_normalized[-seq_length:]

    # Reshape the data for LSTM input (samples, time steps, features)
    last_sequence = last_sequence.reshape((1, seq_length, 1))

    # Predict the next year's production
    prediction = model.predict(last_sequence)

    # Inverse transform the predicted value to the original scale
    predicted_value_original_scale = scaler.inverse_transform(prediction)

    return predicted_value_original_scale[0, 0]

# Create a new DataFrame to store the predicted values
predicted_data = pd.DataFrame(columns=['Provinsi', 'Tahun', 'Produksi'])

# Get unique provinces in the dataset
unique_provinces = data['Provinsi'].unique()

# Predict and add values for each province
for province in unique_provinces:
    province_data = data[data['Provinsi'] == province]['Produksi'].values
    
    # Predict future_steps years into the future
    future_predictions = []
    for _ in range(future_steps):
        predicted_value = predict_next_year(model, scaler, province_data)
        future_predictions.append(predicted_value)
        province_data = np.append(province_data, predicted_value)

    # Add the predicted values to the new DataFrame
    predicted_years = np.arange(data['Tahun'].max() + 1, data['Tahun'].max() + 1 + future_steps)
    predicted_values = np.array(future_predictions)
    
    predicted_data = predicted_data.append(pd.DataFrame({'Provinsi': [province] * future_steps,
                                                        'Tahun': predicted_years,
                                                        'Produksi': predicted_values}),
                                           ignore_index=True)

# Concatenate the original data with the predicted values
data_with_predictions = pd.concat([data, predicted_data], ignore_index=True)

# Display the resulting DataFrame
data_with_predictions
```
